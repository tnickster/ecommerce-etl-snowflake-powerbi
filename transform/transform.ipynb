```python
base_dir = Path().resolve()
raw_dir = base_dir.parent / "extract" / "data"
transform_dir = base_dir / "cleaned"
transform_dir.mkdir(exist_ok=True)
```

```python
for csv_file in raw_dir.glob("*.csv"):
    df = pd.read_csv(csv_file)

    df.columns = (
        df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(r"[^\w_]", "", regex=True)
    )

    df = df.drop_duplicates()

    df = df.dropna(how="all")

    cleaned_name = csv_file.stem + "_clean.csv"
    df.to_csv(transform_dir / cleaned_name, index=False)

```
